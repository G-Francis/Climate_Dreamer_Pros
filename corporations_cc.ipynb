```python
import numpy as np
import pandas as pd
```

```python
corps_cc_df = pd.read_csv('Corporations/Corporations Responses/Climate Change/2018_Full_Climate_Change_Dataset.csv')
na_hq_df = pd.read_csv('Supplementary Data/Locations of Corporations/NA_HQ_public_data.csv')
corps_cc_info = pd.read_csv('Corporations/Corporations Disclosing/Climate Change/2018_Corporates_Disclosing_to_CDP_Climate_Change.csv')
```

```python
# variables to be defined
survey_year = 2018
survey_type = 'Climate Change'
```

```python
corps_cc_df.head()
```

```python
# adding headquarter location data
hq_data = na_hq_df[(na_hq_df['survey_year'] == survey_year) & (na_hq_df['theme'] == survey_type)]
hq_data = hq_data[['account_number','hq_country','address_city','address_state','address_lat','adress_long']]
hq_data = hq_data.rename(columns={'address_city':'hq_city','address_state':'hq_state','address_lat':'hq_lat','adress_long':'hq_long'})

corps_cc_cities = pd.merge(corps_cc_df, hq_data, how='left', on='account_number')
```

```python
corps_cc_cities.organization.unique()
```

```python
corps_cc_info = corps_cc_info[['account_number','activities','sectors','industries','primary_activity','primary_sector','primary_industry','primary_questionnaire_sector']]

corps_cc_full = pd.merge(corps_cc_cities, corps_cc_info, how='left', on='account_number')
```
